# 📚 Week 1: Tensor基础操作

本notebook涵盖：
- Tensor创建与属性
- GPU迁移
- 数学运算
- Reshape与索引

**目标**: 掌握PyTorch核心数据结构操作

```python
import torch
import numpy as np
print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
```

## 1. Tensor创建

```python
# 从Python列表创建
x = torch.tensor([1.0, 2.0, 3.0])
print(f"From list: {x}")

# 固定值
zeros = torch.zeros(3, 4)
ones = torch.ones(2, 3)
print(f"Zeros shape: {zeros.shape}")

# 随机值
rand = torch.randn(2, 3)  # 正态分布 N(0,1)
rand_uniform = torch.rand(3, 3)  # 均匀分布 U(0,1)

# 序列
arange = torch.arange(0, 10, 2)  # 0, 2, 4, 6, 8
linspace = torch.linspace(0, 1, 5)  # 0, 0.25, 0.5, 0.75, 1
print(f"Arange: {arange}")
```

## 2. Tensor属性

```python
x = torch.randn(3, 4)

print(f"Shape: {x.shape}")    # torch.Size([3, 4])
print(f"Dtype: {x.dtype}")   # torch.float32
print(f"Device: {x.device}") # cpu
print(f"ndim: {x.ndim}")     # 2
print(f"numel: {x.numel()}") # 12 (总元素数)
```

## 3. GPU迁移

```python
# 检测GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# 移到GPU
x = torch.randn(3, 4)
x_gpu = x.to(device)
print(f"Tensor device: {x_gpu.device}")
```

## 4. 数学运算

```python
a = torch.tensor([1.0, 2.0, 3.0])
b = torch.tensor([4.0, 5.0, 6.0])

# 基本运算
print(f"a + b = {a + b}")
print(f"a * b = {a * b}")     # 逐元素乘法
print(f"a @ b = {a @ b}")     # 矩阵乘法
print(f"a.sum() = {a.sum()}")
print(f"a.mean() = {a.mean()}")

# 常用函数
print(f"torch.exp(a) = {torch.exp(a)}")
print(f"torch.sqrt(a) = {torch.sqrt(a)}")
print(f"torch.clamp(a, 0, 2) = {torch.clamp(a, 0, 2)}")  # 裁剪到[0,2]
```

## 5. Reshape操作

```python
x = torch.randn(2, 6)
print(f"Original shape: {x.shape}")

# view - 返回新视图（要求连续）
y = x.view(3, 4)
print(f"After view(3,4): {y.shape}")

# reshape - 自动处理连续性
z = x.reshape(12, 1)
print(f"After reshape(12,1): {z.shape}")

# squeeze/unsqueeze
x = torch.randn(1, 3, 1, 4)
print(f"Before squeeze: {x.shape}")
print(f"After squeeze: {x.squeeze().shape}")
print(f"unsqueeze dim=0: {x.unsqueeze(0).shape}")
```

## 6. 索引与切片

```python
x = torch.arange(12).view(3, 4)
print("Matrix x:\n", x)

print(f"x[:, 1] = {x[:, 1]}")       # 第1列
print(f"x[1:, 2:] = {x[1:, 2:]}")   # 第1行后，第2列后
print(f"x[0, 0] = {x[0, 0]}")       # 特定元素

# 布尔索引
mask = x > 5
print(f"Elements > 5: {x[mask]}")
```

## 7. NumPy互转

```python
# Tensor -> NumPy
x = torch.randn(3, 4)
np_x = x.numpy()

# NumPy -> Tensor
np_arr = np.array([[1, 2], [3, 4]])
torch_x = torch.from_numpy(np_arr).float()
```
